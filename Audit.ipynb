```python
import requests

url = "https://data.cityofnewyork.us/resource/erm2-nwe9.csv"
params = {
    "$where": "created_date >= '2025-01-01T00:00:00' AND created_date < '2026-01-01T00:00:00'",
    "$limit": 5000000,
}
r = requests.get(url, params=params)
with open("nyc311_2025.csv", "wb") as f:
    f.write(r.content)
print("done")
```
